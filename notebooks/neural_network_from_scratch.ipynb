```python
class NeuralNetwork:
    def __init__(self, array):
        self.parameters = self.initialize_parameters(array)
        
    def initialize_parameters(self,array):
        params = {}
        L = len(array)
        for l in range(1,L):
            params['W' + str(l)] = (np.ones((array[l-1],array[l])))*0.1
            params['b' + str(l)] = np.zeros((array[l],1))
        return params

    def feed_forward(self,A_prev,w,b):
        Z = np.dot(w.T,A_prev) + b
        return Z

    def L_feed_forward(self,X):
        A = X
        L = len(self.parameters) // 2

        for i in range(1,L+1):
            A_prev = A
            
            Wi = self.parameters['W' + str(i)]
            bi = self.parameters['b' + str(i)]

            A = self.feed_forward(A_prev,Wi,bi)

        return A,A_prev

    def update_parameters(self,y,y_hat,A1,X):
        self.parameters['W2'][0][0] = self.parameters['W2'][0][0] + (0.001 * 2 * (y - y_hat)*A1[0][0])
        self.parameters['W2'][1][0] = self.parameters['W2'][1][0] + (0.001 * 2 * (y - y_hat)*A1[1][0])
        self.parameters['b2'][0][0] = self.parameters['W2'][1][0] + (0.001 * 2 * (y - y_hat))

        self.parameters['W1'][0][0] = self.parameters['W1'][0][0] + (0.001 * 2 * (y - y_hat)*self.parameters['W2'][0][0]*X[0][0])
        self.parameters['W1'][0][1] = self.parameters['W1'][0][1] + (0.001 * 2 * (y - y_hat)*self.parameters['W2'][0][0]*X[1][0])
        self.parameters['b1'][0][0] = self.parameters['b1'][0][0] + (0.001 * 2 * (y - y_hat)*self.parameters['W2'][0][0])

        self.parameters['W1'][1][0] = self.parameters['W1'][1][0] + (0.001 * 2 * (y - y_hat)*self.parameters['W2'][1][0]*X[0][0])
        self.parameters['W1'][1][1] = self.parameters['W1'][1][1] + (0.001 * 2 * (y - y_hat)*self.parameters['W2'][1][0]*X[1][0])
        self.parameters['b1'][1][0] = self.parameters['b1'][1][0] + (0.001 * 2 * (y - y_hat)*self.parameters['W2'][1][0])
        
    def train(self, epochs=100, threshold=0.1):
        prev_loss = float('inf')  # Initialize with a very high value
    
        for i in range(epochs):
            epoch_loss = []
    
            for j in range(df.shape[0]):
                X = df[['cgpa', 'profile_score']].values[j].reshape(2, 1)
                y = df[['lpa']].values[j][0]
    
                y_hat, A1 = self.L_feed_forward(X)
                y_hat = y_hat[0][0]
    
                self.update_parameters(y, y_hat, A1, X)
                epoch_loss.append((y - y_hat) ** 2)
    
            avg_loss = np.mean(epoch_loss)
            print(f'Epoch {i + 1}, Loss: {avg_loss:.4f}')
    
            # Early stopping: if change in loss is less than threshold, stop
            if abs(prev_loss - avg_loss) < threshold:
                print(f"Stopping early at epoch {i + 1}, loss change < {threshold}")
                break
    
            prev_loss = avg_loss  # Update for next iteration
```

```python
obj = NeuralNetwork([2,2,1])
```

```python
X = df[['cgpa', 'profile_score']].values[0].reshape(2,1) # Shape(no of features, no. of training example)
y = df[['lpa']].values[0][0]
obj.train(100)
```
