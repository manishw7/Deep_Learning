```python
class NeuralNetwork:
    def __init__(self,array):
        self.parameters = self.initialize_parameters(array)
        
    def initialize_parameters(self,array):
        parameters = {}
        L = len(array)
        for l in range(1,L):
            parameters['W'+ str(l)] = np.ones((array[l-1],array[l]))*0.1
            parameters['b'+ str(l)] = np.zeros((array[l],1))
            
        return parameters
    
    def feed_forward(self,A_prev,W,b):
        Z = np.dot(W.T,A_prev) + b
        A = self.sigmoid(Z)
        return A
    
    def sigmoid(self,Z):
        A = 1/(1+np.exp(-Z))
        return A
    
    def L_feed_forward(self,X):
        A = X
        L = len(self.parameters) // 2
        
        for l in range(1,L+1):
            
            A_prev = A
            W = self.parameters['W'+ str(l)]
            b = self.parameters['b'+ str(l)]
            
            A = self.feed_forward(A_prev,W,b)
        return A,A_prev
    
    def update_parameters(self,y,y_hat,A_prev,X):
        self.parameters['W2'][0][0] = self.parameters['W2'][0][0] + 0.001 * (y-y_hat) * A_prev[0][0]
        self.parameters['W2'][1][0] = self.parameters['W2'][1][0] + 0.001 * (y-y_hat) * A_prev[1][0]
        self.parameters['b2'][0][0] = self.parameters['b2'][0][0] + 0.001 * (y-y_hat)
        
        self.parameters['W1'][0][0] = self.parameters['W1'][0][0] + (0.0001 * (y - y_hat)*self.parameters['W2'][0][0]*A_prev[0][0]*(1-A_prev[0][0])*X[0][0])
        self.parameters['W1'][0][1] = self.parameters['W1'][0][1] + (0.0001 * (y - y_hat)*self.parameters['W2'][0][0]*A_prev[0][0]*(1-A_prev[0][0])*X[1][0])
        self.parameters['b1'][0][0] = self.parameters['b1'][0][0] + (0.0001 * (y - y_hat)*self.parameters['W2'][0][0]*A_prev[0][0]*(1-A_prev[0][0]))

        self.parameters['W1'][1][0] = self.parameters['W1'][1][0] + (0.0001 * (y - y_hat)*self.parameters['W2'][1][0]*A_prev[1][0]*(1-A_prev[1][0])*X[0][0])
        self.parameters['W1'][1][1] = self.parameters['W1'][1][1] + (0.0001 * (y - y_hat)*self.parameters['W2'][1][0]*A_prev[1][0]*(1-A_prev[1][0])*X[1][0])
        self.parameters['b1'][1][0] = self.parameters['b1'][1][0] + (0.0001 * (y - y_hat)*self.parameters['W2'][1][0]*A_prev[1][0]*(1-A_prev[1][0]))
        
    def train(self,epochs=100,threshold = 0.1):
        prev_loss = float('inf')
        for i in range(epochs):
            epoch_loss = []
            
            for j in range(df.shape[0]):
                X = df[['cgpa','profile_score']].values[j].reshape(2,1)
                y = df[['placed']].values[j][0]
                
                y_hat,A_prev = self.L_feed_forward(X)
                y_hat = y_hat[0][0]
                
                self.update_parameters(y,y_hat,A_prev,X)
                epoch_loss.append(-y*np.log(y_hat) - (1-y)*np.log(1-y_hat))
            avg_loss = np.mean(epoch_loss)
            print(f'Epoch {i + 1}, Loss: {avg_loss:.4f}')
    
            # Early stopping: if change in loss is less than threshold, stop
            if abs(prev_loss - avg_loss) < threshold:
                print(f"Stopping early at epoch {i + 1}, loss change < {threshold}")
                break
    
            prev_loss = avg_loss  # Update for next iteration
```
